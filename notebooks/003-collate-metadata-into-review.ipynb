# Set up

```python
import pandas as pd
```

# Input Arguments

```python
NUM_SAMPLE_BIZ = 400
DATASET_DP = f'../data/yelp_dataset/sample/sample_{NUM_SAMPLE_BIZ}_biz'
```

# Load data

```python
# Load reviews
review_df = pd.read_parquet(f"{DATASET_DP}/yelp_academic_dataset_review_sample_{NUM_SAMPLE_BIZ}_biz.parquet")
```

```python
review_df
```

# Load business

```python
# Load businesses
business_df = pd.read_parquet(f"{DATASET_DP}/yelp_academic_dataset_business_sample_{NUM_SAMPLE_BIZ}_biz.parquet")
business_df
```

# Join review and business

```python
denom_df = (
    pd.merge(
        # TODO: Fix duplicated columns
        review_df.rename(columns={'stars': 'review_stars'}),
        # business_df.rename(columns={'stars': 'business_avg_stars'}),
        business_df.add_prefix("biz_").rename(columns={'biz_business_id': 'business_id'}),
        how='left',
        on='business_id',
        validate='m:1'
    )
)

denom_df
```

```python
denom_df.sample(10000).describe(include='all').T
```

# Persist data

```python
DENOM_REVIEW_FP = f'{DATASET_DP}/denom_review.parquet'
denom_df.to_parquet(DENOM_REVIEW_FP, index=False)
```
