```python
import os

import pandas as pd
```

# Load sample businesses

```python
YELP_DATASET_DP = '../data/yelp_dataset/'
SAMPLE_OUTPUT_DATASET_DP = f'{YELP_DATASET_DP}/sample/sample_100_biz'
os.makedirs(SAMPLE_OUTPUT_DATASET_DP, exist_ok=True)

BUSINESS_DATA_FP = f"{YELP_DATASET_DP}/yelp_academic_dataset_business.json"

NUM_SAMPLE_BUSINESSES = 100
```

```python
business_sample_df = pd.read_json(BUSINESS_DATA_FP, lines=True, nrows=NUM_SAMPLE_BUSINESSES)
```

```python
business_sample_df
```

```python
business_sample_df.describe(include='all').T
```

```python
sample_business_ids = set(business_sample_df['business_id'].unique())
len(sample_business_ids)
```

```python
# Persist
BUSINESS_SAMPLE_DATA_FP = f"{SAMPLE_OUTPUT_DATASET_DP}/yelp_academic_dataset_business_sample_{NUM_SAMPLE_BUSINESSES}_biz.parquet"
business_sample_df.to_parquet(BUSINESS_SAMPLE_DATA_FP, index=False)
```

# Load other data for these sample businesses

## Reviews

```python
REVIEW_DATA_FP = f"{YELP_DATASET_DP}/yelp_academic_dataset_review.json"
```

```python
%%time
review_sample_df = (
    pd.read_json(REVIEW_DATA_FP, lines=True)
    .loc[lambda df: df['business_id'].isin(sample_business_ids)]
)
```

```python
review_sample_df
```

```python
review_sample_df.describe(include='all').T
```

```python
# Persist
REVIEW_SAMPLE_DATA_FP = f"{SAMPLE_OUTPUT_DATASET_DP}/yelp_academic_dataset_review_sample_{NUM_SAMPLE_BUSINESSES}_biz.parquet"
review_sample_df.to_parquet(REVIEW_SAMPLE_DATA_FP, index=False)
```

```python
sample_user_ids = set(review_sample_df['user_id'].unique())
len(sample_user_ids)
```

## Users

```python
USER_DATA_FP = f"{YELP_DATASET_DP}/yelp_academic_dataset_user.json"
```

```python
%%time
user_sample_df = (
    pd.read_json(USER_DATA_FP, lines=True)
    .loc[lambda df: df['user_id'].isin(sample_user_ids)]
)
```

```python
user_sample_df
```

```python
user_sample_df.describe(include='all').T
```

```python
# Persist
USER_SAMPLE_DATA_FP = f"{SAMPLE_OUTPUT_DATASET_DP}/yelp_academic_dataset_user_sample_{NUM_SAMPLE_BUSINESSES}_biz.parquet"
user_sample_df.to_parquet(USER_SAMPLE_DATA_FP, index=False)
```
